In \[19\]:

    # TODO #4: Complete the function.
    def clean_sentence(output):
        sentense = ''
        for i in output:
            word = data_loader.dataset.vocab.idx2word[i]
            if i == 0: continue
            if i == 1: break
            if i == 18: sentense = sentense + word
            else: sentense = sentense + ' ' + word
        return sentense.strip()
